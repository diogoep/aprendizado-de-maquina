```python
model = LinearRegression()
```

```python
X = pd.DataFrame(messi['ano'])
y = pd.DataFrame(messi['gols'])
```

```python
model.fit(X,y)
```

```python
pred_messi = model.predict([[2018]])
```

```python
print(round(pred_messi[0][0],0), target_messi)
```

```python
X = pd.DataFrame(cristiano['ano'])
y = pd.DataFrame(cristiano['gols'])
model.fit(X,y)
pred_cristiano = model.predict([[2018]])
print(round(pred_cristiano[0][0],0), target_cristiano)
```

```python
t = []
t.append([target_cristiano])

r = []
r.append([target_messi])
print(t)
```

```python
print(metrics.mean_absolute_error(pred_cristiano, t))
print(metrics.mean_absolute_error(pred_messi, r))
print(metrics.mean_squared_error(pred_cristiano, t))
print(metrics.mean_squared_error(pred_messi, r))
print(metrics.mean_squared_error(pred_cristiano, t)**0.5)
print(metrics.mean_squared_error(pred_messi, r)**0.5)
```
