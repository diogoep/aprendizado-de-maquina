```python
def corrigirna(data):
    columns = ['alcohol', 'malic_acid','ash','alcalinity_of_ash', 'magnesium','total_phenols','flavanoids','nonflavanoid_phenols','proanthocyanins','color_intensity','hue','od280/od315_of_diluted_wines','proline', 'y']
    for i in columns:
        data[i] = data[i].fillna(value=data[i].mean())
    return data
```

```python
print(data.isna().sum())
print(data.shape[1])
data = corrigirna(data)
```

Por exemplo, se um dado crescer, e tiver uma correlação muito grande com outra coluna, é muito provável que tal dado também sofra modificação.
