```python
import numpy as np
import pandas as pd
import csv
from sklearn.linear_model import LinearRegression
```

```python
treino = pd.read_csv('train.csv')
def f(x):
    if (x==0) or (x==2) :
        return False
    else:
        return True
```

```python
treino
```

```python
teste = pd.read_csv('test.csv')
teste
```

```python
treino = treino[[ 'TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_LC','IN_TREINEIRO','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT']]
treino['NU_NOTA_CN'].fillna(0, inplace = True)
treino['NU_NOTA_CH'].fillna(0, inplace = True)
treino['NU_NOTA_LC'].fillna(0, inplace = True)
treino['NU_NOTA_MT'].fillna(0, inplace = True)
treino['TP_PRESENCA_CN'] = treino['TP_PRESENCA_CN'].apply(f)
treino['TP_PRESENCA_CH'] = treino['TP_PRESENCA_CH'].apply(f)
treino['TP_PRESENCA_LC'] = treino['TP_PRESENCA_LC'].apply(f)
treino['IN_TREINEIRO'] = treino['IN_TREINEIRO'].apply(f)
```

```python

teste['NU_NOTA_CN'].fillna(0, inplace = True)
teste['NU_NOTA_CH'].fillna(0, inplace = True)
teste['NU_NOTA_LC'].fillna(0, inplace = True)

teste['TP_PRESENCA_CN'] = teste['TP_PRESENCA_CN'].apply(f)
teste['TP_PRESENCA_CH'] = teste['TP_PRESENCA_CH'].apply(f)
teste['TP_PRESENCA_LC'] = teste['TP_PRESENCA_LC'].apply(f)
teste['IN_TREINEIRO'] = teste['IN_TREINEIRO'].apply(f)
```

```python
treino_x  = treino.iloc[0:13730,:7]
treino_y  = treino.iloc[0:13730,7]
#treino_x
teste[['TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_LC','IN_TREINEIRO','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC']]
```

```python
lr = LinearRegression()
f = lr.fit(treino_x, treino_y)
predicao_teste = lr.predict(teste[['TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_LC','IN_TREINEIRO','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC']])
predicao_teste_df = pd.DataFrame(predicao_teste)
tabela_final = pd.merge(predicao_teste_df, teste[['NU_INSCRICAO']], on=predicao_teste_df.index)
tabela_final = tabela_final[['NU_INSCRICAO',0]]
tabela_final = tabela_final.rename(columns = {'NU_INSCRICAO':'NU_INSCRICAO', 0 :'NU_NOTA_MT'})
tabela_final['NU_NOTA_MT'] = tabela_final['NU_NOTA_MT'].apply(int)
tabela_final['NU_NOTA_MT'] = tabela_final['NU_NOTA_MT'].apply(abs) 
```

```python
tabela_final.to_csv("output.csv", sep=',',index=False)
```
