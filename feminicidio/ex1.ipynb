```python
import pandas as pd
import matplotlib.pyplot as plt
listaData = []

for ano in range(2016, 2019+1):
    for mes in range(1,12+1):
        filename = ('datasets/DadosBO_' + str(ano)+
                    '_' + str(mes) + '(FEMINICÍDIO).xls')
        try:
            datames = pd.read_csv(filename,sep='\t',encoding='utf-16-le')
            listaData.append(datames)
        except:
            print('erro ' + filename)

data = pd.concat(listaData, sort=True)
data.drop_duplicates(inplace=True, subset='NUMERO_BOLETIM')
```

```python
data['DATAOCORRENCIA'] = pd.to_datetime(data['DATAOCORRENCIA'],format='%d/%m/%Y')
```

## COMO O GRÁFICO ABAIXO ILUSTRA, É FÁCIL NOTAR QUE HÁ UM AUMENTO NA TAXA DE FEMINICÍDIO AO DECORRER DOS ANOS.

```python
%matplotlib inline
data['ANO_BO'].value_counts().plot(kind='bar')
```

## Com relação aos dados presentes no dataframe, é fácil notar que a cidade que possui mais boletins de ocorrência é São Paulo, com uma distância notável dos outros. 

```python
data['CIDADE'].value_counts()[0:20].plot(kind='bar')
```

## Geralmente durante o período da noite é quando mais ocorrem os casos abordados por este dataset.

```python
%matplotlib inline
data['PERIDOOCORRENCIA'].value_counts().plot(kind='bar')
```

## Durante o final de semana é quando mais acontecem os casos abordados neste dataset.

```python
%matplotlib inline
Semana = ['Segunda', 'Terça', 'Quarta','Quinta' , 'Sexta', 'Sabado', 'Domingo']
S = lambda x : Semana[x.dayofweek]
data['DIADASEMANA'] = data['DATAOCORRENCIA'].apply(S)
data['DIADASEMANA'].value_counts().plot(kind='bar')
```

## O período de Maio possui aproximadamente mais de 10 casos que os outros meses, por isso se destaca negativamente, em relação aos outros.

```python
%matplotlib inline
Meses = ['','Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
Mes = lambda x : Meses[x.month]
data['MES'] = data['DATAOCORRENCIA'].apply(Mes)
data['MES'].value_counts().plot(kind='bar')
```
